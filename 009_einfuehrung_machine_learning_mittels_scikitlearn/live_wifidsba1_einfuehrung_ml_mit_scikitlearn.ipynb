## Quiz: Prädiktion für alle Testkunden

Prädiziere nun für alle Testkunden, ob sie jeweils die Versicherung kaufen werden oder nicht!

Problem hierzu zu lösen: Wie kann man über alle Testkunden laufen?

Zusatzaufgabe: Speichere ab, ob die Prädiktion richtig war oder nicht und ermittle dann die sog. Korrekt-Klassifikationsrate (=: wieviel Prozent der Vorhersagen waren richtig)!

Bis 16:10 Uhr!

```python
anz_test_beispiele = len(t_test)

# Laufe über alle Testkunden ...
korrekte_praediktionen = 0
for i in range(0,anz_test_beispiele):
    
    # Prädizierte ob der Testkunde kaufen wird
    pred = praediziere_einen_kunden( t_test.iloc[i] )
    
    # Hole Information, ob der Testkunde wirklich gekauft hat
    gt = t_test.iloc[i]["TravelInsurance"]
    
    # Vergleiche Prädiktion mit Ground Truth
    if pred==gt:
        korrekte_praediktionen += 1
```

## Quiz: KKR bestimmen

Bestimme nun, wie gut der bisher gebaute Klassifikator auf den Testdaten vorhersagen kann, ob eine Kunde eine Reiseversicherung kauft 
oder nicht.

KKR = Korrekt Klassifikations Rate = Wieiviele der Prädiktionen eines Klassifikators sind korrekt?

Hinweise:

- gebe das Prädiktionsergebnis in der Funktion `praediziere_einen_kunden()` auch zurück
- ermittle auch für jeden Testkunden, ob dieser tatsächlich gekauft hat oder nicht

```python
def praediziere_einen_kunden(kunde, merkmal="AnnualIncome"):

    # 1. Einkommen von Testkunde in Variable speichern
    merkmal_testkunde = kunde[merkmal]

    # 2. Suchen in Trainingsdaten nach ähnlichen Kunden
    untere_grenze = 0.90*merkmal_testkunde
    obere_grenze  = 1.00*merkmal_testkunde
    aehnliche_kunden = t_train.loc[ (t_train[merkmal]>=untere_grenze) & (t_train[merkmal]<=obere_grenze) ]
    # t_train.query("AnnualIncome between [untergrenzen, oberegrenze]")?

    # 3. Wieviel Prozent der ähnlichen Kunden haben die
    #    Reiseversicherung gekauft?
    prozent_gekauft = aehnliche_kunden["TravelInsurance"].mean()
    print("\nAnzahl ähnlicher Kunden:", len(aehnliche_kunden), end="   ") 

    # 4. Vorhersage / Prädiktion
    if prozent_gekauft >= 0.5:
        print("Testkunde wird kaufen!")
        return 1
    else:
        print("Testkunde wird nicht kaufen!")
        return 0
    
    
    
anz_test_beispiele = len(t_test)

# Laufe über alle Testkunden ...
korrekte_praediktionen = 0
for i in range(0,anz_test_beispiele):
    
    # Prädizierte ob der Testkunde kaufen wird
    pred = praediziere_einen_kunden( t_test.iloc[i] )
    
    # Hole Information, ob der Testkunde wirklich gekauft hat
    gt = t_test.iloc[i]["TravelInsurance"]
    
    # Vergleiche Prädiktion mit Ground Truth
    if pred==gt:
        korrekte_praediktionen += 1
        
print("KKR=", korrekte_praediktionen / anz_test_beispiele)
```

# Alles in einer Codezelle

```python
# 1. Daten einlesen
import pandas
t = pandas.read_csv("TravelInsurancePrediction.csv")

# 2. Unnötige Spalte(n) wegschmeißen
t = t.drop( 'Unnamed: 0', axis=1 )

# 3. Trennung in Input- und Zielvariabeln
x = t[ ["Age", "AnnualIncome", "ChronicDiseases"] ]
y = t["TravelInsurance"]

# 4. Aufspalten in Trainings- und Testdaten
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)


# 5. Eine eigene Distanzfunktion für die Bestimmung des Abstandes
#    zwischen zwei Kunden definieren
max_alter = x_train["Age"].max()
max_einkommen = x_train["AnnualIncome"].max()

def meine_distanz_funktion(a,b):
    diff_alter     = abs(a[0]-b[0]) / max_alter
    diff_einkommen = abs(a[1]-b[1]) / max_einkommen
    diff_krankheit = abs(a[2]-b[2])
    
    distanz = diff_alter + diff_einkommen + diff_krankheit
    return distanz


# 6. k-NN Klassifikator lernen
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=15, weights="uniform", metric=meine_distanz_funktion)
model.fit(x_train, y_train)

# 7. Modellgüte bestimmen
from sklearn.metrics import accuracy_score
pred = model.predict(x_test)
accuracy_score(pred, y_test)
```

# Testen verschiedener ML-Verfahren auf dem gleichen Problem

```python
import pandas

from sklearn.model_selection import cross_val_score

from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier


# 1. Daten einlesen
t = pandas.read_csv("TravelInsurancePrediction.csv")

# 2. Unnötige Spalte(n) wegschmeißen
t = t.drop( 'Unnamed: 0', axis=1 )

# 3. Wandle nicht-numerische Spalten (kategorialen Spalten)
#    in numerische Spalten um über das Prinzip der one-hot-Enkodierung
t = pandas.get_dummies(t)

# 4. Nehme alle Merkmale außer "TravelInsurance" als Inputs
x = t.drop("TravelInsurance", axis=1)
y = t["TravelInsurance"]


# 5. Definiere die ML-Verfahren, die wir alle gegeneinander antretren lassen wollen
verfahrens_liste = [ "KNeighborsClassifier()",
                     "DecisionTreeClassifier()", 
                     "GradientBoostingClassifier()",
                     "SVC()",
                     "RandomForestClassifier(n_estimators=1000)"]

# 6. Teste jedes Verfahren einzelnt
for verfahren in verfahrens_liste:
    
    # Modell erzeugen
    model = eval(verfahren)
    
    # Modellgüte bestimmen
    scores = cross_val_score(model, x,y, cv=5)
    
    print( f"Verfahren: {verfahren} --> scores={scores} --> Mittel={scores.mean():.2f}" )
```
